# Digit Classification, ANN version

## Import

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot
import seaborn as sns

from sklearn import metrics
from sklearn.preprocessing import Normalizer

from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, BatchNormalization
from keras.utils.np_utils import to_categorical
from keras.callbacks import EarlyStopping
```

## Récupération des données

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
print('X_train: ' + str(X_train.shape))
print('y_train: ' + str(y_train.shape))
print('X_test:  '  + str(X_test.shape))
print('y_test:  '  + str(y_test.shape))
```

## Analyse des données

```python
sns.countplot(y_train).set(title='Répartition des labels de train')
```

```python
sns.countplot(y_test).set(title='Répartition des labels de test')
```

On observe une répartition équivalente et homogène entre les données de train et de test.

```python
for i in range(9):  
    pyplot.subplot(330 + 1 + i)
    pyplot.imshow(X_test[i], cmap=pyplot.get_cmap('gray'))
pyplot.show()
```

## Préparation des données

```python
X_train = X_train.reshape((X_train.shape[0], 28*28)).astype('float32')
X_test = X_test.reshape((X_test.shape[0], 28*28)).astype('float32')
```

```python
print('X_train: ' + str(X_train.shape))
print('y_train: ' + str(y_train.shape))
print('X_test:  '  + str(X_test.shape))
print('y_test:  '  + str(y_test.shape))
```

```python
X_train = X_train / 255.0
X_test = X_test / 255.0
```

Une fois les données normalisé et reshape elles sont prètes à être utilisé

```python
y_train = to_categorical(y_train, num_classes = 10)
y_test = to_categorical(y_test, num_classes = 10)
```

## Création des modèles

### Model 1

```python
model = Sequential()
model.add(Dense(32, input_dim = 28 * 28, activation= 'relu'))
model.add(Dense(64, activation = 'relu'))
model.add(Dense(10, activation = 'softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=10, batch_size=100, verbose=1)
```

```python
_, train_acc = model.evaluate(X_train, y_train, verbose=0)
_, test_acc = model.evaluate(X_test, y_test, verbose=0)
print('Train: %.3f, Test: %.3f' % (train_acc, test_acc))
```

```python
# plot loss during training
pyplot.subplot(211)
pyplot.title('Loss')
pyplot.plot(history.history['loss'], label='train')
pyplot.plot(history.history['val_loss'], label='test')
pyplot.legend()
# plot accuracy during training
pyplot.subplot(212)
pyplot.title('Accuracy')
pyplot.plot(history.history['accuracy'], label='train')
pyplot.plot(history.history['val_accuracy'], label='test')
pyplot.legend()
pyplot.show()
```
